```python
import requests
from bs4 import BeautifulSoup

levelHerfs = [] #10個level的頁面
pages = [] #每一個level的所有頁面
movies = [] #每一個頁面的電影網址
infos = [] #個別電影資訊集合
```

```python
for i in range(1,7):
    levelHerf = "https://movies.yahoo.com.tw/moviegenre_result.html?genre_id="+str(i)
    #print(levelHerf)
    levelHerfs.append(levelHerf)

print(levelHerfs)

"""
url = 'https://movies.yahoo.com.tw/moviegenre_result.html?genre_id=1'
r = requests.get(url)
soup = BeautifulSoup(r.text)
"""
```

```python
#動作片所有頁面
pages.append('https://movies.yahoo.com.tw/moviegenre_result.html?genre_id=1')
for i in range(2,128):
    page = "https://movies.yahoo.com.tw/moviegenre_result.html?genre_id=1&page="+str(i)
    #print(page)
    pages.append(page)
    
pages = set(pages)
pages
```

```python
#動作片所有頁面電影網址
for page in pages:
    url = page
    r = requests.get(url)
    soup = BeautifulSoup(r.text)

    for d in soup.find_all('div', class_="release_foto"):
        movie = d.find('a', class_='gabtn')['href']
        movies.append(movie)
    
movies = set(movies)
movies
```

```python
infos = []
for movie in movies:
    level = []
    info = {}
    url = movie
    r = requests.get(url)
    soup = BeautifulSoup(r.text)
    
    info['中文名稱'] = soup.find('div', class_='movie_intro_info_r').find('h1').text
    info['英文名稱'] = soup.find('div', class_='movie_intro_info_r').find('h3').text
    for d in soup.find_all('div', class_="level_name"):
        if d.find('a', class_="gabtn") == None:
            continue
        else:
            level.append(d.find('a', class_="gabtn")['data-ga'][23:25])
    info['分類'] = level
    info['劇情介紹'] = soup.find('div', class_='gray_infobox_inner').find('span').text.strip()
    info['上映日期'] = soup.find('div', class_='movie_intro_info_r').find('span').text[5:]
    infos.append(info)

infos 
```

```python
#CSV
import pandas as pd

df = pd.DataFrame(infos)
df.to_csv('movies.csv', encoding='utf_8_sig')

df
```
